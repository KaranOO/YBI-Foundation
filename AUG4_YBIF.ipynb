#Data Visualisation

```python
df = pd.read_csv('https://raw.githubusercontent.com/ybifoundation/Dataset/main/Titanic.csv')
```

#Detecting NULL data

```python
df.info()
```

#Setting 'sex' column for data visualization

```python
df.replace({'sex':{'female':0,'male':1}},inplace=True)
```

#Replacing the empty data

```python
df['age'].fillna(0,inplace=True)
df['fare'].fillna(0,inplace=True)
df['body'].fillna(0,inplace=True)
```

```python
df
```

```python
fig, ax = plt.subplots(figsize=(10,10))
plt.title("AGE CHART")
#ax.set_label("AGE CHART")
#ax.pie(df['age'])
ax.hist(df['age'],color='orange')
ax.set_xlabel('Age',size=20)
ax.set_ylabel('Number of people',size=20)
ax.set_xlim(0,100)
ax.set_ylim(0,350)
plt.show()
```

```python
x = df['sex'].value_counts()
labels = 'Male','Female'
explode=(0,0.1)
fig, ax = plt.subplots(figsize=(10,10))
plt.title("Total females")
ax.pie(x,labels=labels,startangle=240,explode=explode,autopct='%1.1f%%',shadow=True,textprops={'fontsize': 14})
plt.show()
```

```python
sm = len(df[( df['survived']==1 ) & ( df['sex']==1 )])   #survived male
sf = len(df[( df['survived']==1 ) & ( df['sex']==0 )])   #survived female
l = len(df[df['survived']==0])    #left
x = sm ,sf , l
labels = 'survived Male','survived Female','left'
explode=(0.1,0,0)
fig, ax = plt.subplots(figsize=(10,10))
plt.title("Total males survived",size=20)
ax.pie(x, labels=labels, explode=explode, autopct='%1.1f%%',shadow=True,textprops={'fontsize': 14})
plt.show()
```
